```python
import pandas as pd
```

```python
def load_df(path, k):
    dfs = []
    for i in range(1, k):
        p = path.format(i)
        df = pd.read_csv(p)
        dfs.append(df)

    dfs_a = [v['A'].values for v in dfs]
    return pd.DataFrame(dfs_a).transpose()
```

```python
base = load_df("outputs/submission_bert_768_seqlen256-L{}.csv", 11)
big = load_df("outputs/submission_bert_1024_seqlen256-L{}.csv", 24)
```

```python
base
```

```python
corrs = pd.concat([big, base], axis=1, keys=['df1', 'df2']).corr().loc['df2', 'df1']
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 15))
layers = [i for i in range(1, 24)]
x_layers = [i for i in range(2, 12)]
sns.heatmap(corrs, annot=True, fmt=".2f", xticklabels=layers, yticklabels=x_layers)
```
