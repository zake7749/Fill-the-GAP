```python
import pandas as pd
```

```python
dfs = []
for i in range(2, 24):
    path = "outputs/submission_bert_1024_seqlen256-L{}.csv".format(i)
    df = pd.read_csv(path)
    dfs.append(df)
```

```python
dfs_a = [v['A'].values for v in dfs]
```

```python
cor_a = pd.DataFrame(dfs_a)
```

```python
cor_a.transpose().corr()
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15, 15))
layers = [i for i in range(2, 24)]
sns.heatmap(cor_a.transpose().corr(), annot=True, fmt=".2f", xticklabels=layers, yticklabels=layers)
```
